```python
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data['sex'] = data['sex'].astype('category').cat.codes
data['smoker'] = data['smoker'].astype('category').cat.codes
data['day'] = data['day'].astype('category').cat.codes
data['time'] = data['time'].astype('category').cat.codes
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2, random_state=123)
```

```python
regression_setup = setup(data=train,
    target='tip',
    session_id=123,
    normalize=True,
    polynomial_features=True,
    remove_multicollinearity=True,
    log_experiment=False, # Disable experiment logging
    experiment_name='tips_regression',
    verbose=False )
# Suppress logs)
```

```python
best_model = compare_models()
```

```python
print (best_model)
```

```python
tuned_model = tune_model(best_model, optimize = 'MAE')
```

```python
evaluate_model(tuned_model)
```

```python
final_model = finalize_model(tuned_model)
predictions = predict_model(final_model, data=test)
```

```python
from  sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
true_values= test['tip']
print (predictions.columns)
```

```python
predicted_values = predictions['prediction_label']
```

```python
mse = mean_squared_error(true_values, predicted_values)
r2 = r2_score(true_values, predicted_values)
print("Mean Squared Error: {mse}".format(mse=mse))
print("R-squared: {r2}".format(r2=r2))
```

```python
import numpy as np
```

```python
df = sns.load_dataset("tips")
x = df[["total_bill", "size"]]
y = df["tip"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(f"Mean Squared Error: {round(mean_squared_error(y_test, y_pred), 4)}")
print(f"R2 Score: {round(r2_score(y_test, y_pred), 4)}")
print(f"Mean Absolute Error: {round(mean_absolute_error(y_test, y_pred), 4)}")
# root mean squared error
print(f"Root Mean Squared Error: {round(np.sqrt(mean_squared_error(y_test, y_pred)), 4)}")
```

```python
data = sns.load_dataset('titanic')
```

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
data['age'] = imputer.fit_transform(data[['age']])
data.isnull().sum().sort_values(ascending=False)
```

```python
data[['deck', 'embarked', 'embark_town']].dtypes
```

```python
# impute the deck, embarked, and embark_town columns with the most frequent value
data['deck'].fillna(data['deck'].mode()[0], inplace=True)
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)
data['embark_town'].fillna(data['embark_town'].mode()[0], inplace=True)
data.isnull().sum().sort_values(ascending=False)
```
